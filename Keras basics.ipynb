```python
import tensorflow as tf
```

```python
from sklearn.datasets import load_iris
```

```python
X = load_iris()['data']
target = load_iris()['target']
```

```python
X
```

```python
target
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot = OneHotEncoder()
```

```python
from tensorflow.keras.utils import to_categorical
```

```python
y = to_categorical(target)
```

```python
y
```

```python
from sklearn.preprocessing import OneHotEncoder
one_hot = OneHotEncoder()
```

```python
y1 = one_hot.fit_transform(target.reshape(-1, 1))
y1
```

```python
y1.toarray()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scalar = MinMaxScaler()
```

```python
scaled_X_train = scalar.fit_transform(X_train)
```

```python
scaled_X_test = scalar.transform(X_test)
```

```python
scaled_X_train
```

```python
scaled_X_test
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(8, input_dim = 4, activation = 'relu'))
model.add(Dense(8, input_dim = 4, activation = 'relu'))
model.add(Dense(3, activation = "softmax"))
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(scaled_X_train, y_train, epochs=150)
```

```python
y_pred = model.predict(scaled_X_test)
```

```python
y_pred_classes = y_pred.argmax(axis = 1)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
cm = confusion_matrix(y_test.argmax(axis = 1), y_pred_classes)
cm
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test.argmax(axis = 1), y_pred_classes))
```
