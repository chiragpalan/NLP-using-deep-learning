### Function for preprocessing text

#### Reading file as text string

```python
def read_file(file_path):
    with open(file_path) as f:
        str_text = f.read()
    return str_text
```

```python
read_file('moby_dick_four_chapters.txt')
```

### Tokenizing and removing stop words

```python
import spacy
nlp = spacy.load('en_core_web_lg', disable = ['parser', 'tagger', 'ner'])
nlp.max_length = 1198623
```

```python
# function to remove punctuation 
# nlp('My name is chirag') --> doc object, tokens of sentances looks like list
def separate_punc(doc_text):
    return [token.text.lower() for token in nlp(doc_text) if token.text not in '\n\n \n\n\n!"-#$%&()--.*+,-/:;<=>?@[\\]^_`{|}~\t\n ']
```

```python
# cvonverting mobi dick text into tokens...
# ...and removing specific punctuations of mobi dick

d = read_file('moby_dick_four_chapters.txt')

tokens = separate_punc(d)
```

### Creating sequence of tokens

```python
# Creating sequence of 25 training words and next word is to be predicted i.e. 26 words in total
train_len = 25 + 1
# list to hold text sequence of 26 tokens
text_sequence = []

for i in range(train_len, len(tokens)):
    # Grabing sequence of 26 tokens one by one by shifting 1 step
    seq = tokens[i - train_len:i] # for first iteration it will grab 0 to 25 tokens and so on...
    
    # Adding seq to list
    text_sequence.append(seq)
```

### Converting string tokens into numbers using keras for feeding into NN

#### Keras tokenization

```python
from tensorflow.keras.preprocessing.text import Tokenizer
```

```python
# creating instance of Tokenizer
tokenizer = Tokenizer() 
# Assigning unique number to each word
tokenizer.fit_on_texts(text_sequence)
# making 
sequences = tokenizer.texts_to_sequences(text_sequence)
```

### Creatin gLSTM based model

```python
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense, LSTM, Embedding
```

### Creating sequential model

```python
# from tensorflow.keras.losses import SparseCategoricalCrossentropy
# loss = SparseCategoricalCrossentropy()
def creat_model(vocabulary_size, seq_len):
    model = Sequential()
    model.add(Embedding(vocabulary_size, 25, input_length = seq_len))
    model.add(LSTM(150, return_sequences = True))
    model.add(LSTM(150))
    model.add(Dense(150, activation = 'relu'))
    
    model.add(Dense(vocabulary_size, activation = 'softmax'))
    
    model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
    
    model.summary()
    
    return model
```

```python
# converting labels to one hot encoding using keras
from tensorflow.keras.utils import to_categorical
```

### Generating new text

```python
from random import randint
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

```python
def generate_text(model, tokenizer, seq_len, seed_text, num_gen_words):
    # final output
    output_text = []
    
    # initial seed sequence
    input_text = seed_text
    
    for i in range(num_gen_words):
        
        # take the input text string and encod it to a sequence
        encoded_text = tokenizer.texts_to_sequences([input_text])[0]
        
        # pad sequences to our trained rate
        pad_encoded = pad_sequences([encoded_text], maxlen=seq_len, truncating='pre')
        
        # predict class probability of each word
        pred_word_ind = model.predict(pad_encoded, verbose = 0)[0].argmax()
        
        # Grab word
        pred_word = tokenizer.index_word[pred_word_ind]
        
        # update sequence of input text (shifting one over with new word)
        input_text = tokenizer.index_word[pred_word_ind]
        
        output_text.append(pred_word)
    
    return " ".join(output_text)
```
